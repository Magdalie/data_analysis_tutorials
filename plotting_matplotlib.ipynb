```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots() # axes is container for visualisation, figure can have more than one axes
print(type(fig))
print(type(axes))

# Adding Data to an Axes
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 23, 45, 33, 22, 45, 55]

axes.plot(x,y)
```

```python
import pandas as pd

df = pd.DataFrame.from_dict({
    'Year': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Computer Sales': [12500, 13000, 13500, 15000, 14000, 16000, 17000, 18000, 16500, 17000, 19000],
    'TV Sales': [13000, 20000, 18000, 19000, 19500, 21000, 23000, 24000, 22000, 22500, 25000]
})
print(df.head())
fig, ax = plt.subplots()
ax.plot(df['Year'], df['Computer Sales'], label='Computer Sales', color='r', linestyle='dotted')
ax.plot(df['Year'], df['TV Sales'], label='TV Sales', color='aquamarine', linestyle='-.')
# Styling Matplotlib Titles
ax.set_title('Computer and TV Sales Over Time', size=18, fontweight='bold')
ax.set_xlabel('Year', style='italic')
ax.set_ylabel('# of Sales', style='italic')
ax.legend()
ax.set_ylim(bottom=0)
```

```python
df = pd.DataFrame.from_dict({
    'Year': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Computer Sales': [12500, 13000, 13500, 15000, 14000, 16000, 17000, 18000, 16500, 17000, 19000],
    'TV Sales': [13000, 20000, 18000, 19000, 19500, 21000, 23000, 24000, 22000, 22500, 25000]
})
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.plot(df['Year'], df['Computer Sales'], label="Computer Sales")
ax.plot(df['Year'], df['TV Sales'], label="TV Sales")
ax.set_title("TV and computer sales over time")
ax.set_xlabel("Year")
ax.set_ylabel("# sold")
ax.set_ylim(bottom=0)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.pie([df['Computer Sales'].sum(), df['TV Sales'].sum()], labels=['Computer Sales', 'TV Sales'], autopct='%1.1f%%')
ax.set_title('Computer and TV Sales Compared')
```

```python
# bar plot
fig, ax = plt.subplots()
width=0.4 # must be less than 1
ax.bar(df['Year'], df['TV Sales'], width=width, label= "TV Sales")
ax.bar(df['Year']+width, df['Computer Sales'], width=width, label="Computer Sales")

ax.set_title("Sales")
ax.legend()
```
