```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('penguins')
df.head()
```

```python
# A highlight of the parameters of the sns.relplot() function
sns.relplot(
    x=None,             # The variable to use as the x-axis
    y=None,             # The variable to use as the y-axis
    hue=None,           # A grouping variable to use to color data points
    size=None,          # A grouping variable to define the size of data points
    data=None,          # The input data structure, such as a DataFrame
    legend='auto',      # How to draw the legend
    kind='scatter',     # Either 'scatter' or 'line'
)
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x = 'flipper_length_mm', 
                y='body_mass_g', 
                data=df,
                hue='species',
                style='sex')
plt.title("Body mass vs Flipper length")
plt.xlabel('Flipper length [mm]')
plt.ylabel("Body mass [g]")
```

```python
sns.relplot(x='body_mass_g', y='flipper_length_mm', data=df, hue='sex', col='species', col_wrap=2)
plt.title("Body mass vs Flipper length")
plt.xlabel('Flipper length [mm]')
plt.ylabel("Body mass [g]")
```

```python
sns.catplot(data=df, x='species', y='body_mass_g',kind='bar')
plt.title("Bodymass by species")
```

```python
sns.set_palette('pastel')
sns.catplot(data=df, 
            x='species', 
            y='body_mass_g',
            kind='bar',
            hue='sex')
plt.title("Bodymass by species")
```

```python
# color palletes
pallete_pastel = sns.color_palette('pastel')
sns.palplot(pallete_pastel)
```

```python
sns.histplot(data=df,
             x='bill_length_mm',
             bins=10)
```

```python
sns.displot(data=df,
             x='bill_length_mm',
             bins=10,
             col='species') # displot is more generalized method than histplot; with kind='hist' it is the same plot
```
